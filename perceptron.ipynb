```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.random.rand(input_size + 1)  # +1 for bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        x = np.insert(x, 0, 1)  # Add bias input (1)
        return self.activation(np.dot(self.weights, x))

    def train(self, X, y):
        X = np.insert(X, 0, 1, axis=1)  # Add bias to all inputs
        for epoch in range(self.epochs):
            print(f"Epoch {epoch +1}/{self.epochs}")  # Print current epoch
            for i in range(len(y)):
                y_pred = self.activation(np.dot(self.weights, X[i]))
                error = y[i] - y_pred
                self.weights += self.learning_rate * error * X[i]

# Define input and output for AND, OR, XOR gates
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
AND_targets = np.array([0, 0, 0, 1])
OR_targets = np.array([0, 1, 1, 1])
XOR_targets = np.array([0, 1, 1, 0])

# Train Perceptron for AND gate
print("Perceptron for AND gate")
pcn_and = Perceptron(input_size=2)
pcn_and.train(inputs, AND_targets)
for i in inputs:
    print(f"Input {i} => Output {pcn_and.predict(i)}")

# Train Perceptron for OR gate
print("\nPerceptron for OR gate")
pcn_or = Perceptron(input_size=2)
pcn_or.train(inputs, OR_targets)
for i in inputs:
    print(f"Input {i} => Output {pcn_or.predict(i)}")

# Train Perceptron for XOR gate (will fail)
print("\nPerceptron for XOR gate (will fail)")
pcn_xor = Perceptron(input_size=2)
pcn_xor.train(inputs, XOR_targets)
for i in inputs:
    print(f"Input {i} => Output {pcn_xor.predict(i)}")
```
